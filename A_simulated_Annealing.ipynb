## 3.2. The A* Algorithm

<font size="4"> The A* algorithm is a type of Informed search algorithms; it uses heuristics to guide the search. A* algorithm uses an evaluation function f(n) which estimates the total cost path of the solution, it consists of the heuristic function h(n), and the function g(n) which calculates the exact path cost from the initial state to the current state, as illustrated: 
</font><br>
<font size="4">
<font color = "green">$$f(n) = h(n) + g(n)$$ </font>
</font>
<font size="4">The optimality of A* algorithm depends on the implementation the developer chooses. In the case of using tree-search, the heuristic must be admissible which means it must be less than or equal to the real path cost from that state to the goal state h* (n), as follow h(n) ≤ h*(n). In the other hand, if the graph-search is used the heuristic must be consistent which means the heuristic function of the current state must be less than or equal to the heuristic function of the successor state plus the step cost between them c(n,a,n’), as follow h(n) ≤ c(n,a,n’) + h(n’). As illustrated in Figure.2.
    <br>In our problem, we have chosen to implement the algorithm using the tree-search strategy, so the heuristic must be admissible since we want to reach an optimal solution. A straight-line distance is always the shortest distance which is proven mathematically, so we have used it as a heuristic.
</font>

2.3. خوارزمية A*

خوارزمية A* نوع من خوارزميات البحث الموجَه، أي تستخدم التوقعات لتوجيه البحث. خوارزمية A* تستخدم دالة التقدير f(n) التي تقدر قيمة الطريق الكلية للحل وتتكون من دالة التوقع h(n) ودالة g(n) التي تحسب القيمة الفعلية للطريق من الحالة الابتدائية إلى الحالة الحالية، كما هو موضح:<br></font></h2>
<font size="4"><font color = "green">$$f(n) = h(n) + g(n)$$ </font></font>
<h2 dir="rtl"><font size="3">خوارزمية A* المثلى تعتمد على اختيار المطور لطريقة تنفيذها. في حالة استخدام البحث الشجري، التوقع يجب أن يكون مقبولاً بحيث يكون أقل أو مساوي للقيمة الفعلية للطريق من أي حالة إلى حالة الهدف h*(n)، أي h(n) ≤ h*(n).  من جهة أخرى إذا كان البحث البياني هو المستخدم فيجب أن يكون التوقع متوافق أي أن دالة التوقع للحالة الحالية يجب أن تكون أقل أو مساوية لدالة التوقع للحالة التالية بالإضافة إلى قيمة الخطوة بينهما c(n,a,n’)، كالتالي h(n) ≤ c(n,a,n’) + h(n’). وكما هو موضح في الشكل.2.<br>
في مشكلتنا تم اختيار أسلوب البحث الشجري لتنفيذ الخوارزمية، لذلك قيمة التوقعات يجب أن تكون مقبولة حتى نصل للحل الأمثل. البراهين الرياضية تثبت أن مسافة الخط المستقيم دائماً ما تكون أقصر مسافة لذا تم استخدامها كقيمة للتوقع.

### 3.2.1. The Properties of A* Algorithm
<font size="4"> A* algorithm is complete which means it will always find a solution if one exists, also it is optimal if the heuristic is admissible or consistent depend on the strategy developer use as we have explained before. The time complexity in A* algorithm is O($b^m$) and the space complexity is O($b^m$) where m is the remaining steps to the goal and b is the branching factor. 
</font>

### 3.2.2. Implementation
<font size="4"> The A* algorithm starts with an initial state, it inserts the state to the fringe. After removing the front node from the fringe checks if it is the goal state, in case it is a goal it returns a solution, otherwise, it will insert all its successors to the fringe and sort them according to the value of their evaluation functions. Then, it will continue iterating through the fringe till it finds a solution or it will return failure when the fringe is empty.

بعد فهم مفاهيم خوارزمية A*، هنا سنقوم بتطبيقها على المشكلة باستخدام لغة Python البرمجية.

```python
class Node:
    def __init__(self, state, parent, h, step_cost):
        self.state = state
        self.parent = parent
        self.h = h
        if parent :
            self.g = parent.g + step_cost 
        else: 
            self.g = step_cost
        self.f = self.g + h
```

<font size="4">class Node contains only a constructor to create a node that has the following attributes:<br>
state: which represents the city name.<br>
parent: which refers to the previous city lead to this current city, it is None when it is the initial state.<br>
h: is the heuristic of the city to the goal.<br>
g: is the real cost path from the initial state to this current state, calculated by summing the g(parent) and step cost between them, unless it is the initial state it will be equal to 0.<br>
f: is the evaluation function (summation of g and h).<br>
</font>
<h2 dir="rtl">
<font size="3">
class Node يحتوي فقط على مكوِّن لصنع عقدة تمتلك الخواص التالية:<br>
الحالة: التي تمثل اسم المدينة.<br>
الأب: يشير إلى المدينة السابقة التي تؤدي إلى هذه المدينة, تكون قيمته None حين يكون هو الحالة الابتدائية.<br>
h: التوقع للمدينة إلى حالة الهدف.<br>
g: القيمة الفعلية للطريق من الحالة الابتدائية إلى هذه الحالة، تحسب عن طريق جمع قيمة g للأب و قيمة الخطوة بينهما, وإن كانت الحالة الابتدائية ستكون مساوية للصفر.<br>
f: الدالة التقديرية (مجموع h وg).
<br>
</font>
</h2>

```python
def make_node(state, parent_node): # -> node
    node = Node(state,parent_node, heuristic[state], distance[parent_node.state][state])
    return node
```

<font size="4"> make_node method has two parameters, state ,and parent node. it creates node by invoking the constructor of class Node and pass the state, parent_node, the heuristic of the state from the heuristic dictionary, and the step cost from the distance dictionary. then it will return the created node to the caller.</font>
<h2 dir="rtl">
<font size="3">
دالة make_node لها متغيرين، الحالة وعقدة الأب. تصنع عقدة من خلال استدعاء المكوِّن لـ class Node وتمرر الحالة، عقدة الأب، التوقع من قاموس التوقعات، قيمة الخطوة من قاموس المسافة. ثم تعيد العقدة المنشأة إلى المنادي.
<br>
</font>
</h2>

```python
def expand(node): # -> children:
    children = []
    for i in distance[node.state]:
        child = make_node(i, node)
        children.append(child)
    return children
```

<font size="4">expand method takes a node as a parameter to iterate over its successors and add them to children list and return the children list.</font>
<h2 dir="rtl">
<font size="3">

```python
def solution(node) : #-> path:
    path = node.state
    cost = node.g
    node = node.parent
    while node != None:
        path =  node.state + "->" + path
        node = node.parent
    path = path + "\t cost: " + str(cost)
    return path
```

<font size="4">solution method takes a node as a parameter to track the path taken from this node to the initial node. While it is backtracking, it appends the state of each node to the path. In the end, it will append also the cost of that path and then return the path.</font>
<h2 dir="rtl">
<font size="3">
دالة solution تأخذ عقدة كمتغير لتتبع الطريق المتَّخذ من هذه العقدة إلى العقدة الابتدائية. بينما تقوم بالتتبع الخلفي تضيف الحالة لأي عقدة لمتغير الطريق. في النهاية تضيف قيمة الطريق إلى نفس المتغير ثم تعيد متغير الطريق.<br>

```python
def A_star (init_node):
    fringe=[init_node]
    count = 0
    while (fringe) :
        curr = fringe.pop(0)
        if (curr.h == 0):
            return solution(curr), count
        count += 1
        fringe.extend(expand(curr))  
        fringe.sort(key=lambda x: x.f)
    return "failure"
```

<font size="4">A_star method takes the initial node as a parameter and returns a solution or failure; it inserts the initial node in the fringe and loop through the fringe if it is empty it will return "failure". Otherwise, it will remove the front of the fringe then it checks the node if it is the goal state it will invoke solution method to return the path (solution), in case it is not the goal state it will insert all its successors to the fringe by invoking expand method. At the end of each iteration, the fringe will be sorted by the nodes evaluation function (f) thus node with smallest f will be at the front.<br>
count variable is used to count the number of expanded nodes whether it is within the soultion or not.<br>
</font>

## 4.2. The Simulated Annealing algorithm
<br>
<font size="4">
    To anneal the metal, expose the metal to high temperature at the beginning to change its physical properties. Then, try to cool the temperature slowly enough while forming the final structure of the metal before it completely cools down.<br>
A simulated annealing algorithm is inspired by the process of annealing the metal. to form the final solution of the problem, it starts with a high temperature where the 
algorithm can pick up moves even if they are bad moves. However, cooling the temperature down will decrease the probability of picking bad moves with time, so we need to decrease the temperature slowly enough to give it a better chance to find a solution.<br>
Simulated annealing algorithm has an advantage over the hill-climbing algorithm, where hill climbing may sometimes not find a solution because of its continuous looking for a next better move therefore stuck in a local optimum. However, simulated annealing solves this problem by picking bad moves according to the temperature degree, which means taking actions may not improve the current state but making the search jumps over the local optima and tries more solutions. <br>

</font>

```python
def random_child(node): # ->  child:
    list_keys = [ k for k in distance[node.state]]
    r = random.choice(list_keys)
    return S_make_node(r, node)
```

<font size="4">

    دالة random_child تأخذ العقدة كمدخل، ثم تقوم بتخزين كل أبناء هذه العقدة في قائمة، بعد ذلك تقوم باختيار واحد من هؤلاء الأبناء عشوائياَ وتقوم بإعادة الابن المختار.

```python
def S_solution(node) : #-> path:
    if not node.parent:
        return node.state + ", cost: 0"
    path = node.state
    cost = distance[node.parent.state][node.state]
    node = node.parent
    while node != None and node.parent != None:
        path =  node.state + "->" + path
        cost += distance[node.parent.state][node.state]
        node = node.parent
    path = node.state + "->" + path +", cost: " + str(cost)
    return path  
```

<font size="4">
    S_solution method here takes a node considered as a solution, then it will backtrack the path to the initial node storing the path and cost of the solution, then it returns them.<br>
</font>

 At each iteration the temperature will decrease according to the schedule method, then it will stop iterating if it encounters a goal or when the temperature is lower than 1 (cool). <br>
 otherwise, it will select a random successor then it will compute E as a heuristic of(current) – heuristic of(random successor). in case E is positive, the randomly selected successor has a lower heuristic (good move) so it will be chosen as a next current node immediately. in the other case where E is negative, the randomly selected successor has a higher heuristic (bad move), thus bad move will be accepted regarding the probability value $𝑒^{𝐸/𝑇}$. when the temperature is high the probability is high so it is more likely to be greater than the random number generated (acceptance of bad move is high), and as the temperature cools down the probability will decrease and it will be less likely to be greater than the random number generated (acceptance of bad move is low).<br>

    دالة  simulated_annealing_Tree تستقبل العقدة الابتدائية كمدخل ثم ستقوم بالدوران مع الملاحظة أن الوقت يبدأ بصفر ويتزايد في كل دورة.<br>
في كل دورة تنقص درجة الحرارة بقيمة معينة مأخوذة من دالة schedule . سيتم إيقاف الدوران في حال الوصول للهدف، أو في حال كانت درجة الحرارة قد أصبحت أقل من واحد (باردة).<br>
    في حال عدم تحقق أي من الشرطين سيتم اختيار ابن عشوائي للعقدة الحالية ثم نقوم بحساب E كـ (توقع العقدة الحالية – توقّع العقدة الابن العشوائي ). اذا كانت قيمة E موجبة هذا يعني أن توقع العقدة الأبن أقل من توقع العقدة الحالية والعقدة الابن أقرب للهدف (حركة جيدة)، إذاَ سيتم اختيار هذا الابن العشوائي كعقدة حاليّة على الفور، أما في حال كانت قيمة E سالبة فهذا يعني أن قيمة توقّع العقدة الابن أكبر من قيمة توقّع العقدة الحالية أي أن العقدة الابن أبعد عن الهدف (حركة سيئة)  وسيتم قبول هذه الحركة السيئة حسب قيمة الاحتمالية $𝑒^{𝐸/𝑇}$. فعند درجات الحرارة العالية تكون الاحتمالية عالية، إذن سيكون هنالك فرصة أعلى لكون الاحتمالية أكبر من عدد عشوائي بين الصفر وواحد وستكون فرصة قبول الحركة السيئة أعلى، مع مرور الوقت تبدأ درجة الحرارة بالتناقص مما ينقص قيمة الاحتمالية بالتالي تقل فرصة كون قيمة الاحتمالية أكبر من العدد العشوائي، أي أن فرصة قبول الحركة السيئة ستكون أقل.<br>
    متغير count يستخدم لحساب عدد العقد الموسعة سواء كانت من ضمن الحل أو لا. 
</font><br>
</h2>

```python
def simulated_annealing_Graph (init_node):
    curr = init_node
    closed = [curr.state]
    count = 0
    for time in range(999999):
        T = schedule(time)
        if  (curr.h == 0) | (T < 1.0):
            return S_solution(curr), count
        next_suc = random_child(curr)
        count += 1
        if closed.__contains__(next_suc.state):
            continue    
        E = curr.h - next_suc.h
        if E > 0 :
            curr = next_suc
            closed.append(curr.state)
        else:
            p = math.exp(E/T)
            if p > random.random():
                curr = next_suc    
                closed.append(curr.state)
```

<font size="4">
    simulated_annealing_Graph method is as simulated_annealing_Tree method but the difference is that nodes have been explored will be stored in a closed list, therefore any node in this list will be explored only once.
    
</font>

<h2 dir="rtl">
<font size="3">
    دالة simulated_annealing_Graph كـدالة simulated_annealing_Tree لكن الفرق بينهما أن دالة simulated_annealing_Graph يقوم بتخزين جميع العقد التي قام بتوسيعها في قائمة تسمى القائمة المغلقة، بهدف عدم توسيع العقدة والبحث فيها لأكثر من مرة. 
</font><br>
</h2>

```python
def schedule(time):
    T = 1000
    cooling_rate = 0.003
    T = T * ((1 - cooling_rate)**time)
    return T.__round__(2)
```

<font size="4">
    schedule method takes time as a parameter, then calculate the temperature(T) according to the factor of cooling_rate and time. T is set to  1000 now as an initial value, and cooling_rate is the rate at which T decreases and set to 0.003 as a constant.
if you increase the cooling_rate, T will decrease rapidly that will make the search stops early.
</font><br>
<h2 dir="rtl">
<font size="3">
    دالة schedule تقوم بأخذ الوقت كمدخل لها، ثم تقوم بحساب درجة الحرارة T معتمدة على عامل معّدل التبريد والوقت. تم اختيار القيمة 1000 كقيمة ابتدائية لدرجة الحرارة كما أن قيمة معدّل التبرد تم اختيارها بقيمة 0.003 كقيمة ثابتة. 
إذا قمت بزيادة معدل التبريد ستتناقص قيمة درجة الحرارة بمعدل أسرع، مما يجعل البحث يتوقف بوقت أبكر.
